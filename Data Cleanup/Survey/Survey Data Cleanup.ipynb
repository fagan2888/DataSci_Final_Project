```python
import pandas as pd
import numpy as np
```

```python
Survey_data = pd.read_csv('Alcohol Consumption Survey.csv')
Cleaned_data = pd.read_csv('CleanData.csv')
```

```python
#assign integers to values
Survey_data['school_num'] = 2
Survey_data = Survey_data.rename(columns={'What is your age?':'age'})
Survey_data['sex_num'] = Survey_data['What is your gender?'].map({'Female':0, 'Male':1})
Survey_data['address_num'] = Survey_data['What area do you live in?'].map({'Urban':0, 'Suburban':0, 'Rural':1})
Survey_data['famsize_num'] = Survey_data['How large is the family you live with?'].map({'Less than or equal to three':0, 'Do not live with family':0 ,'Greater than three':1})
Survey_data['reason_num'] = Survey_data['Why did you choose the University of Utah?'].map({'Close to home':0 ,'School Reputation':1 ,'Program Preference':2, 'Other':3})
Survey_data['traveltime'] = Survey_data['What is your home to school commute time?'].map({'Less than 15 min':1, '15 min to 30 min':2, '30 min to 1 hour':3, 'Greater than 1 hour':4})
Survey_data['studytime'] = Survey_data['How many hours do you study a week for all of your classes?'].map({'Less than 2 hours':1 ,'2 to 5 hours':2 ,'5 to 10 hours':3, 'Greater than 10 hours':4})
Survey_data['failures'] = Survey_data['How many classes have you failed in the past during your time at the University of Utah?'].map({'0':0, 'None':0, 'none':0 ,'1':1, '2':2, '3':3, '4':4, '5':4})
Survey_data['famsup_num'] = Survey_data['Do you receive educational support from your family?'].map({'Yes':0, 'No':1})
Survey_data['internet_num'] = Survey_data['Do you have internet access at home?'].map({'Yes':0, 'No':1})
Survey_data['romantic_num'] = Survey_data['Are you in a romantic relationship?'].map({'Yes':0, 'No':1})
Survey_data['freetime'] = Survey_data['How much free time do you have outside of school?'].map({'1- very low amount':1 ,'2 - low amount':2, '3 - some amount':3, '4 - high amount':4, '5 - very high amount':5})
Survey_data['Dalc'] = Survey_data['How much alcohol do you consume on an average weekday?'].map({'1 - very low amount':1 ,'2 - low amount':2, '3 - some amount':3, '4 - high amount':4, '5 - very high amount':5})
Survey_data['Walc'] = Survey_data['How much alcohol do you consume on an average weekend?'].map({'1 - very low amount':1 ,'2 - low amount':2, '3 - some amount':3, '4 - high amount':4, '5 - very high amount':5})
Survey_data['health'] = Survey_data['How is your current health?'].map({'1 - very bad':1, '2 - bad':2, '3 - okay':3, '4 - good':4, '5 - very good':5})
Survey_data['religious'] = Survey_data['Are you religious?'].map({'Yes':0, 'No':1})

#should tutoring data be assigned to 'paid'?
Survey_data['paid_num'] = Survey_data['Do you receive tutoring for any of your classes?'].map({'Yes':0, 'No':1})
```

```python
#drop nan
Survey_data = Survey_data.dropna()

Survey_data.describe()
```
