```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (5,5)
plt.style.use('ggplot')

Cleaned_data_pie = pd.read_csv("CleanData.csv")
Survey_data_pie = pd.read_csv('SurveyData.csv')
```

```python
#Rename columns
Survey_data_pie = Survey_data_pie.rename(columns={'What is your gender?':'Gender'})
Cleaned_data_pie['Gender'] = Cleaned_data_pie['sex'].map({'F':'Female', 'M':'Male'})
```

```python
#Gender pie charts
Survey_data_pie['Gender'].value_counts().sort_index().plot(kind='pie', autopct='%1.1f%%')
plt.title('Gender, U of U Survey')
plt.show()

Cleaned_data_pie['Gender'].value_counts().sort_index().plot(kind='pie', autopct='%1.1f%%')
plt.title('Gender, Portugal data')
plt.show()
```

```python
#Rename columns
Survey_data_pie = Survey_data_pie.rename(columns={'How many hours do you study a week for all of your classes?':'Studytime per week'})
Cleaned_data_pie['Studytime per week'] = Cleaned_data_pie['studytime'].map({1:'Less than 2 hours', 2:'2 to 5 hours', 3:'5 to 10 hours', 4:'Greater than 10 hours'})
```

```python
#Studytime pie charts
Survey_data_pie['Studytime per week'].value_counts().sort_index().plot(kind='pie', autopct='%1.1f%%')
plt.title('Studytime per week, U of U Survey')
plt.show()

Cleaned_data_pie['Studytime per week'].value_counts().sort_index().plot(kind='pie', autopct='%1.1f%%')
plt.title('Studytime per week, Portugal data')
plt.show()
```

```python
#Rename columns
Survey_data_pie = Survey_data_pie.rename(columns={'Are you in a romantic relationship?':'Romantic Relationship'})
Cleaned_data_pie['Romantic Relationship'] = Cleaned_data_pie['romantic'].map({'yes':'Yes', 'no':'No'})
```

```python
#Romantic pie charts
Survey_data_pie['Romantic Relationship'].value_counts().sort_index().plot(kind='pie', autopct='%1.1f%%')
plt.title('Romantic Relationship, U of U Survey')
plt.show()

Cleaned_data_pie['Romantic Relationship'].value_counts().sort_index().plot(kind='pie', autopct='%1.1f%%')
plt.title('Romantic Relationship, Portugal data')
plt.show()
```
